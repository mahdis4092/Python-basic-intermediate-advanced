If Else statement

```python
age = 16
if age>=18:
    print('You are eligible for vote')
else:
    print("You are not eligible for vote")
```

if elif else statment

```python
likes = 200
require = 200
if likes> require:
    print("Enough great outcome")
elif likes==require:
    print("Got Attention")
else:
    print("Need to improve")
```

```python
#while
i = 1
while i<= 10:
    print("Iteration: ",i)
    i+=1
```
